```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('../datasets/titanic.csv')
df
```

```python
df.isnull()
```

```python
df.fillna('NA',inplace=True)
```

**Set PassengerId as the index.**

```python
df.set_index("PassengerId", inplace = True)
df.head()
```

**Create a pie chart presenting the male/female as proportions**

```python
gender=df.groupby('Sex')['Sex'].value_counts()
gender.plot(kind='pie',title='Male/Female Proportions',figsize=(4, 4),autopct='%1.0f%%',colors=['yellow','orange'],shadow=True)
```

**Create a scatterplot with the Fare payed and the Age, differ the plot color by gender**

```python
plt.figure() 
category1 = df[df.Sex=='male'].plot.scatter('Age', 'Fare', color='blue',label='male') 
df[df.Sex=='female'].plot.scatter('Age', 'Fare',color='red',label='female',ax=category1)
```

**How many people survived?**

```python
df['Survived'].loc[df['Survived']==1].count()
```

**Create a histogram with the Fare payed**

```python
df.hist(column='Fare')
```
