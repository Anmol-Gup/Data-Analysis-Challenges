```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('../datasets/tips.csv')
df
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

**Delete the Unnamed 0 column**

```python
df.drop('Unnamed: 0',inplace=True,axis=1)
df
```

**Plot the total_bill column histogram**

```python
sns.histplot(df['total_bill'],color='y',kde=True,bins=[0,10,20,30,40,50])
plt.title('Total Bill Distribution')
plt.show()
```

**Create a scatter plot presenting the relationship between total_bill and tip**

```python
sns.scatterplot(x=df['total_bill'],y=df['tip'],hue=df['sex'],style=df['sex'],size=df['sex'],sizes=(30,40))
plt.title('Total Bill Vs Tip')
plt.show()
```

**Create one image with the relationship of total_bill, tip and size.**

```python
#plt.figure(figsize=(16,9))
sns.scatterplot(df,x='total_bill',y='tip',hue='size',size='size',style='size',palette='deep')
plt.title('Distribution of total bill & tip among size')
plt.show()
```

**Present the relationship between days and total_bill value**

```python
sns.barplot(x='total_bill',y='day',data=df,palette='hot')
plt.title('Distribution of Total Bills by Day')
plt.show()
```

**Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex**

```python
sns.scatterplot(df,x='tip',y='day',hue='sex',style='sex')
plt.title('Tip Amount by Day and Sex')
plt.xlabel('Tip')
plt.ylabel('Day')
plt.show()
```

**Create a box plot presenting the total_bill per day differentiation the time (Dinner or Lunch)**

```python
sns.boxplot(x=df['total_bill'],y=df['day'],hue=df['time'])
plt.show()
```

**Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.**

```python
sns.histplot(df,x='tip',hue='time',multiple='dodge',bins=5)
plt.title('Distribution of Tip by Time')
plt.show()
```

**Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker**

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
sns.scatterplot(df[df.sex=='Male'],x='total_bill',y='tip',hue='smoker',style='smoker',palette='magma')
plt.title('Total Bill Vs Tip for Male by Smoker')
#plt.show()
plt.subplot(1,2,2)
sns.scatterplot(df[df.sex=='Female'],x='total_bill',y='tip',hue='smoker',style='smoker',palette='hot')
plt.title('Total Bill Vs Tip for Female by Smoker')
plt.show()
```
