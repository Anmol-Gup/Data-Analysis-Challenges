```python
import pandas as pd
import numpy as np
```

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data' 
wine = pd.read_csv(url) 
wine
```

```python
# checking for null values.
wine.isnull()
```

**Delete the first, fourth, seventh, nineth, eleventh, thirteenth and fourteenth columns.**

```python
wine=wine.set_axis(np.arange(len(wine.columns)), axis="columns")
wine.drop([0,3,6,8,10,12,13],axis=1,inplace=True)
```

```python
wine
```

**Assign the columns as below: alcohol, malic_acid, alcalinity_of_ash, magnesium, flavanoids, proanthocyanins, hue**

```python
wine.columns=['alcohol', 'malic_acid', 'alcalinity_of_ash', 'magnesium', 'flavanoids', 'proanthocyanins', 'hue']
wine
```

**Set the values of the first 3 rows in the alcohol column as NaN**

```python
wine.loc[:2,'alcohol']=pd.NA
wine.loc[:2,'alcohol']
```

**Now set the value of the rows 3 and 4 of the magnesium column as NaN**

```python
wine.loc[2:3,'magnesium']=pd.NA
wine.loc[2:3,'magnesium']
```

**Fill in the null values (NaN) with the number 10 in the alcohol column and 100 in magnesium column.**

```python
wine['alcohol'].fillna(10,inplace=True)
wine['magnesium'].fillna(100,inplace=True)
wine
```

**Count the number of missing values in the entire dataset.**

```python
wine.isnull().sum().sum(),wine.isna().sum().sum(),wine.isnull().values.sum()
```
